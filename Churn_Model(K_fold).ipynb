# Part 1 - Data Preprocessing

```python
# Importing the libraries
import numpy as np #importing the required libraries
import pandas as pd #importing the required libraries
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv') #reading the dataset
X = dataset.iloc[:, 3:13] #data slicing (selects all rows in dataset and columns 4 through 13)
y = dataset.iloc[:, -1] #data slicing (select till second last columm)

X = X.values
y = y.values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder #importing the required functions from library
labelencoder_X_1 = LabelEncoder() #creating an instance of the LabelEncoder class (used to transform categorical variables into numerical variables)
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])

labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html
from sklearn.compose import ColumnTransformer #importing the required functions from library
ct = ColumnTransformer([("Country", OneHotEncoder(),[1])], remainder="passthrough") # The last arg ([0]) is the list of columns you want to transform in this step
X = ct.fit_transform(X)
X = X[:, 1:]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split #importing the required functions from library
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler #importing the required functions from library
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Part 2 - Now let's make the ANN!

```python
# Importing the Keras libraries and packages
import tensorflow as tf #importing the library
print ("TensorFlow version: " + tf.__version__) #rinting TF version
```

```python
# Evaluating the ANN
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier #importing required functions from imported libraries
from sklearn.model_selection import cross_val_score #importing required functions from imported libraries
from tensorflow.keras.models import Sequential #importing required functions from imported libraries
from tensorflow.keras.layers import Dense #importing required functions from imported libraries
```

```python
def build_classifier():
    classifier = Sequential() #creating a sequential model
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11)) #creating a dense layer with 6 neurons, and uniform as weight intializer ,relu as an activation function and with 11 neurons as input
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu')) #creating a dense layer with 6 neurons, and uniform as weight intializer ,relu as an activation function
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid')) #creating a dense layer with 1 neurons, and uniform as weight intializer ,sigmoid as an activation function
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) #compiling the model with adam optimizer, and binart entropy a loss function, accuracy class (Calculates how often predictions equal labels)
    return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100, verbose=1) # building the model, having 10 as batch size (10 samples per batch) and running for 100 epochs)
# n_jobs = -1 can cause errors if using tensorflow gpu for multiprocessing
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10) #having 10 class validation folds (traiing data to split in 10 parts), and having-1 as no. of parrelle proccesing CPU (-1 indicates that all available CPU cores will be used for maximum efficiency.
mean = accuracies.mean() #calculating the mean accuracy of the model
variance = accuracies.std()
```
